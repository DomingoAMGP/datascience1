Influencia en las Actividades Diarias
La integración del celular en la rutina cotidiana es casi total. Ha redefinido conceptos básicos como la atención y la inmediatez. Desde la planificación de rutas con GPS y la gestión bancaria, hasta el seguimiento de la salud con aplicaciones de fitness, la dependencia del móvil ha automatizado y simplificado innumerables tareas. Esto trae consigo beneficios claros en eficiencia y acceso a la información, pero también un alto costo en términos de atención y presencia. La necesidad constante de revisar notificaciones interrumpe ciclos de trabajo profundo y descanso, fomentando un estado de multitarea continua que, paradójicamente, puede reducir la productividad general y la calidad de las interacciones cara a cara.

Por un lado, el acceso instantáneo a Internet, bibliotecas digitales y herramientas colaborativas lo convierte en un recurso de aprendizaje inigualable. En conclusión, la relación entre el celular y el ser humano es íntima e inevitablemente compleja. 1.¿Quienes usan mas el celular hombres o mujeres?

```
male_female = dfphone[dfphone['Gender']!='Other'].groupby('Gender')['Daily_Usage_Hours'].mean()
male_female
```

2.¿Que edad usa mas el celular?

```
edad = dfphone.groupby('Age')['Daily_Usage_Hours'].mean()
edad
```

```
grade = dfphone.groupby('School_Grade')['Daily_Usage_Hours'].mean()
grade
```

```
plt.style.use('ggplot')
fig, ax = plt.subplots(1, 4, figsize=(15,4))

genero.plot(kind='bar',ax=ax[0],xlabel='Género', ylabel='Horas', title='Uso del celular por género ')

male_female.plot(kind='bar',ax=ax[1],xlabel='SIS-Género', ylabel='Horas', title='Uso del celular por género', color='blue')

edad.plot(kind='pie',ax=ax[2],xlabel='Edad', ylabel='Horas', title='Uso del celular por edad ', legend =True)

grade.plot(kind='line',ax=ax[3],xlabel='Grado', ylabel='Horas', title='Uso del celular por grado ', legend =True)
```

Podemos concluir que el uso del celular no esta influenciado por la edado genero. Se muestra una tendencia poco siginificativa de usar menos el celular en grados mayores.

3.¿Influye positiva o negativamente en el resultado academico el uso del celular?

```
academico= dfphone.groupby('Academic_Performance')['Daily_Usage_Hours'].mean()
academico_formato = academico.round(2)
academico_formato.head()
```

```
fig, ax = plt.subplots(1, 2, figsize=(15,4))

academico_formato.plot(kind='line',ax=ax[0], xlabel='Performance académico', ylabel='# de horas en el celular', title='Relación entre el uso del celular y el grado académico', legend =True)

#Para ser mas precisos y poder realizar una conclusión veremos la diferencia con un gráfico donde las horas de uso de celular sean en base a las de estudio

academico2 = dfphone.groupby('Academic_Performance')['Time_on_Education'].mean()
academico_formato2 = academico2.round(2)

academico_formato2.plot(kind='line', ax=ax[1], xlabel='Performance académico', ylabel='# de horas en el celular educativo', title='Relación entre el uso del celular (educativo) y el grado académico ')
```

# Seaborn

4.   ¿Cuál es la relación entre la ansiedad de los estudiantes y el nivel de adicción?

```
fig, ax = plt.subplots(figsize=(20,10))
sns.scatterplot(
    data=dfphone,
    x='Addiction_Level',
    y='Anxiety_Level',
    alpha=0.7                 # Transparencia
)
plt.title("Relación Ansiedad vs Adiccion")
plt.show()
```

```
fig, ax = plt.subplots(1,2 ,figsize=(10,10))
orden_deseado = ['7th','8th','9th','10th','11th','12th']
dfphone['School_Grade'] = pd.Categorical(dfphone['School_Grade'], categories=orden_deseado, ordered=True)

ax[0].set_title('Nivel de ansiedad debido al uso de celular por Grado')
ax[0].set_ylim(0,10)
sns.lineplot(
  data = dfphone,
  x='School_Grade',
  y='Anxiety_Level',
  ax=ax[0],
  #markers=True,
  errorbar=None,
  legend=True)

ax[1].set_title('Nivel de adicción al celular por Grado')
ax[1].set_ylim(0,10)
dfphone['School_Grade'] = pd.Categorical(dfphone['School_Grade'], categories=orden_deseado, ordered=True)
sns.lineplot(
  data = dfphone,
  x='School_Grade',
  y='Addiction_Level',
  ax=ax[1],
  #markers=True,
  errorbar=None,
  legend=True)
```

El nivel de ansiedad y adiccion son directamente proporcionales. A mayor adiccion mayor ansiedad

5.¿En qué proporción de tiempo utilizan el celular los estudiantes?. 2 graficos, 1  de linea matplot y otro de barras con seaborn

```
academico = dfphone.groupby('Academic_Performance')['Daily_Usage_Hours'].mean()
academico_formato = academico.round(2)
#academico_formato

fig, ax = plt.subplots(figsize=(10,10))

academico_formato.plot(kind='line', xlabel='Performance académico', ylabel='# de horas en el celular', title='Relación entre el uso del celular y el grado académico')

academico2 = dfphone[dfphone['Phone_Usage_Purpose'] == 'Education'].groupby('Academic_Performance')['Time_on_Education'].mean()
academico_formato2 = academico2.round(2)

academico2.plot(kind='line', xlabel='Performance académico', ylabel='# de horas en el celular', title='Relación entre el uso del celular (educativo) y el grado académico ')

academico3 = dfphone[dfphone['Phone_Usage_Purpose'] == 'Gaming'].groupby('Academic_Performance')['Time_on_Gaming'].mean()
academico_formato3 = academico3.round(2)

academico3.plot(kind='line', xlabel='Performance académico', ylabel='# de horas en el celular', title='Relación entre el uso del celular (Juegos) y el grado académico ')

academico4 = dfphone[dfphone['Phone_Usage_Purpose'] == 'Social Media'].groupby('Academic_Performance')['Time_on_Social_Media'].mean()
academico_formato4 = academico4.round(2)

academico4.plot(kind='line', xlabel='Performance académico', ylabel='# de horas en el celular', title='Relación entre el uso del celular (Redes Sociales) y el grado académico ')
ax.legend(['Todos', 'Educativo', 'Juegos', 'Redes Sociales'])
plt.title('Horas de uso del celular VS Performance académico' )
plt.show()
```

```
data = dfphone[(dfphone['Phone_Usage_Purpose']=='Education') |
               (dfphone['Phone_Usage_Purpose']=='Social Media') |
               (dfphone['Phone_Usage_Purpose']=='Gaming')
               ]
min_cal = data['Academic_Performance'].min()
max_cal = data['Academic_Performance'].max()
bins = np.arange(start=min_cal, stop=max_cal + 10, step=10)

data['Rango'] = pd.cut(data['Academic_Performance'], bins=bins, right=False) # 'right=False' excluye el límite superior
promedio_por_rango = data.groupby('Rango')['Daily_Usage_Hours'].mean()
promedio_por_rango
fig, ax = plt.subplots(figsize=(10,10))
sns.barplot(
  data = data ,
  x='Rango',
  y='Daily_Usage_Hours',
  hue='Phone_Usage_Purpose',
  #markers=True,
  errorbar=None)



```

Se puede concluir que los alumnos utilizan el celular 5horas en promedio, siendo las redes sociales un 50% de este un 35% dedicado a los juegos y un 15% dedicado al estudio.
